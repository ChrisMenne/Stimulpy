```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, solve, pi

# Desired area
area = 100

# Calculate side of a square with the same area
side_square = float(area ** 0.5)

# Calculate radius of a circle with the same area
radius_circle = float((area / pi) ** 0.5)

# Calculate side of an equilateral triangle with the same area
s = symbols('s')
side_triangle = float(solve((s**2 * (3**0.5) / 4) - area, s)[0])

# Function to plot square
def plot_square(ax, side,fill=0,col='r'):
    square = plt.Rectangle((-side/2, -side/2), side, side, fill=fill, color=col)
    ax.add_patch(square)

# Function to plot circle
def plot_circle(ax, radius,fill=0,col='g'):
    circle = plt.Circle((0, 0), radius, fill=fill,color=col)
    ax.add_patch(circle)

# Function to plot equilateral triangle
def plot_triangle(ax, side,fill=0,col='b'):
    height = (side**2 - (side/2)**2)**0.5
    triangle = plt.Polygon([(-side/2, -height/3), (side/2, -height/3), (0, 2*height/3)], fill=fill, color=col)
    ax.add_patch(triangle)

```

```python
# Plot the shapes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
plot_square(axs[0], side_square,fill=1)
axs[0].set_title('Square')

plot_circle(axs[1],radius_circle,fill=0)
axs[1].set_title('Circle')

plot_triangle(axs[2], side_triangle)
axs[2].set_title('Equilateral Triangle')

for ax in axs:
    ax.set_xlim(-10,10)
    ax.set_ylim(-10,10)
    ax.set_aspect('equal')
    ax.grid(False)
    ax.get_yaxis().set_visible(False)
    ax.get_xaxis().set_visible(False)

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np
import math

#show empty frame 

def standard_ax(ax=ax,line_width=3):
    """
    
    
    """
    ax.set_xlim([-10,10])
    ax.set_ylim([-10,10])
    ax.grid(True)
    ax.get_yaxis().set_visible(False)
    ax.get_xaxis().set_visible(False)
    ax.spines['left'].set_linewidth(line_width)
    ax.spines['bottom'].set_linewidth(line_width)
    ax.spines['top'].set_linewidth(line_width)
    ax.spines['right'].set_linewidth(line_width)
    pass

def empty_plot(line_width=3):
    """
    
    
    
    """
    fig,ax = plt.subplots(1,1,figsize=(10,10))
    standard_ax(ax)
    plt.show()
    pass

#define polygon function square

def square_stim(color="r",line_width=3,edge=None):
    """
    
    
    """
    #get points set up
    x = [-8.5,-8.5,8.5,8.5]
    y = [-8.5,8.5,8.5,-8.5]
    fig , ax = plt.subplots(1,1,figsize=(10,10))
    ax.fill(x,y,color,edgecolor=edge,linewidth=10)
    standard_ax(ax,line_width=3)
    plt.show()
    pass

#define polygon function triangle

def triangle_stim(color="r",line_width=3,edge='white'):
    """
    
    
    """
    #get points set up
    x = [-8.5,8.5,0]
    y = [-math.sqrt(216.75)/2,-math.sqrt(216.75)/2,math.sqrt(216.75)/2]
    fig , ax = plt.subplots(1,1,figsize=(10,10))
    ax.fill(x,y,color,edgecolor=edge,linewidth=10)
    standard_ax(ax,line_width=3)
    plt.show()
    pass

#define circle polygon
def circle_stim(color="r",line_width=3,edge='white'):
    """
    
    
    """
    #get points set up
    radius = 8.5
    fig , ax = plt.subplots(1,1,figsize=(10,10))
    circle = plt.Circle((0, 0), radius,facecolor=color,edgecolor=edge,linewidth=10)
    ax.add_patch(circle)
    standard_ax(ax,line_width=3)
    plt.show()
    pass

#define star polygon
def star_points(center, radius, num_points):
        points = []
        angle = 2 * np.pi / num_points
        start_angle = np.pi / 2  # Start from the top point

        for i in range(2 * num_points):
            r = radius if i % 2 == 0 else radius / 2
            theta = start_angle + i * angle / 2
            x = center[0] + r * np.cos(theta)
            y = center[1] + r * np.sin(theta)
            points.append([x, y])
        
        return points

def star_stim(color="r",line_width=3,edge='white'):
    """
    
    
    """
    # Define the function to calculate star vertices
    # Calculate points
    points=star_points((0,-0.8555),8.94,5)
    # Create the plot
    fig, ax = plt.subplots(figsize=(10,10))

    # Create the polygon (star) and add it to the plot
    star = Polygon(points, closed=True,edgecolor=edge, facecolor=color,linewidth=10)
    ax.add_patch(star)
    standard_ax(ax)

    # Display the plot
    plt.show()
```
