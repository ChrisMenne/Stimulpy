```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, solve, pi

# Desired area
area = 100

# Calculate side of a square with the same area
side_square = float(area ** 0.5)

# Calculate radius of a circle with the same area
radius_circle = float((area / pi) ** 0.5)

# Calculate side of an equilateral triangle with the same area
s = symbols('s')
side_triangle = float(solve((s**2 * (3**0.5) / 4) - area, s)[0])

# Function to plot square
def plot_square(ax, side,fill=0,col='r'):
    square = plt.Rectangle((-side/2, -side/2), side, side, fill=fill, color=col)
    ax.add_patch(square)

# Function to plot circle
def plot_circle(ax, radius,fill=0,col='g'):
    circle = plt.Circle((0, 0), radius, fill=fill,color=col)
    ax.add_patch(circle)

# Function to plot equilateral triangle
def plot_triangle(ax, side,fill=0,col='b'):
    height = (side**2 - (side/2)**2)**0.5
    triangle = plt.Polygon([(-side/2, -height/3), (side/2, -height/3), (0, 2*height/3)], fill=fill, color=col)
    ax.add_patch(triangle)

```

```python
# Plot the shapes
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
plot_square(axs[0], side_square,fill=1)
axs[0].set_title('Square')

plot_circle(axs[1],radius_circle,fill=0)
axs[1].set_title('Circle')

plot_triangle(axs[2], side_triangle)
axs[2].set_title('Equilateral Triangle')

for ax in axs:
    ax.set_xlim(-10,10)
    ax.set_ylim(-10,10)
    ax.set_aspect('equal')
    ax.grid(False)
    ax.get_yaxis().set_visible(False)
    ax.get_xaxis().set_visible(False)

plt.show()
```
